```python
# cycle olarak itertools kullandık, bunun yerine counter tanımlayıp bunu da kullanabilirdik. Değindiğimiz yanlış yere oynama durumunu ise -1 ekleyerek çözebiliriz.
from itertools import cycle


# boardı oluşturuyoruz. yazdığımız oyun n x n boardlarda çalışacak şekilde
game_board = [[0, 0, 0],
              [0, 0, 0],
              [0, 0, 0]
              ]

# genel bir kazanma durumu tanımladık. Buarada liste'nin ilk elemanının sayısını listenin eleman sayısına eşit olup olmadığını kontrol ettik. Başka bir ifade ile listenin bütün elemanlarının aynı olup olmadığını
def win(l):
    if l.count(l[0]) == len(l) and l[0] != 0:
        return True
    else:
        return False

# yatay olarak kazanma durumunu kontrol ediyoruz. Listedeki elementlerin her bir elementlerini tek tek kontrol ederek. 
def horizontal_win(game_board):
    for i in game_board:
        horizontals = []
        for j in i:
            horizontals.append(j)
        if win(horizontals):
            return True 

# dikey kazanma durumuna bakıyoruz. ilk indexi sabit tutarak diğer index'i değiştiriyoruz ve bu şekilde iterasyona devam ediyoruz
def vertical_win(game_board):
    for i in range(len(game_board)):
        verticals = []
        
        for j in range(len(game_board)):
                verticals.append(game_board[j][i])
        if win(verticals):
            return True 

# sol üstten sağ alta çapraz kazanma durumu. Indexler aynı şekilde hareket edeceği için bunu kolaylıkla gerçekleştirebiliyoruz.
def first_diagonal(game_board):
    dia_win = []
    for i in range(len(game_board)):
        dia_win.append(game_board[i][i])
    if win(dia_win):
        return True

# sağ üstten sol alta çapraz kazanma durumu. Biri artarken diğeri azalan indexleri kullanarak yapıyoruz.
def second_diagonal(game_board):
    dia_win = []
    for i, j in zip(range(len(game_board)), range(len(game_board)-1, -1, -1)):
        dia_win.append(game_board[i][j])

    if win(dia_win):
        return True 

# berabere kalma durumuna bakıyoruz. Boardın içindeki elementlerin tek tek içinde 0 olup olmadığını kontrol ediyoruz.
def tie(game_board):
    zeros = 0
    for i in game_board:
            zeros += i.count(0)
    if not zeros:
        return True

# 2 oyuncumuz var ve cycle kullanıyoruz. 1. satırda açıklandığı gibi
players = [1, 2]
players = cycle(players)
winner = False 

# bir kazanan olmadığında oyun devam etmesi için winner değişkenimizi False olarak atıyor ve bu şekilde loopa oluşturuyoruz.
while not winner:
    for i in game_board:
        for j in i:
            print(j, end='    ')
        print('\n')
    # terminalde daha güzel durması için boşluklar koyduk.
    
    current_player = next(players) # her aşamada oyuncuyu değiştiriyoruz 1 ve 2 olarak
    print(current_player, 'is playing') # hangi oyuncunun oynadığına dair bilgilendirme yapıyoruz.
    row = int(input('row? ')) # hangi row'a oynamak istediğini,
    column = int(input('column? ')) # hangi column'a oynamak istediğini soruyoruz.

    if game_board[row][column] == 0: # oynanmak istenen elementin yerinin 0 olup olmadığını kontrol ediyoruz.
        game_board[row][column] = current_player # eğer 0'sa bunu o anki oyuncu ile değiştiriyoruz.
    else:
        print('Not playable, try again')
        current_player = next(players) # burada daha önceden oynanan bir yere oynanmasını önlüyor ve böyle bir durumla karşılaştığımızda oyuncu tekrar oynayabilsin diye tekrar next yapıyoruz.
    
    if first_diagonal(game_board) or second_diagonal(game_board) or horizontal_win(game_board) or vertical_win(game_board): # herhangi bir win durumu olup olmadığını kontrol ediyoruz.
        winner = True # loopumuzu devam ettiren winner değişkeni True olarak değiştiriyoruz ki oyun devam dursun
        print(current_player, 'wins')

    if tie(game_board):
        print('tiieeee!!!')
        winner = True # 87. satırla aynı mantıkla bunu değiştirdik. Yukarıda tanımladığımız berabera olma durumunu kullandık.

```
